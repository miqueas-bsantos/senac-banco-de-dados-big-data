Criar classe de conexão com Banco de dados

```python
class Connection():
    """
        Classe que agrupa as funções de 
        gerenciamento do Banco de Dados. 
    """
    
    def __init__(self, user, password, host, port, database, engine):
        """
            Classe que instancia as conexões com 
            o serviço PostgreSql.

            PARAMETROS:
                user: Username do PostgreSql;
                password: Senha do PostgreSql;
                host: Ponto de acesso do PostgreSql;
                port: Porta do ponto de acesso do PostgreSql;
                database: Banco de dados do PostgreSql.
        """

        self.connection = f"{engine}://{user}:{password}@{host}:{port}/{database}"
        self.engine = create_engine(
            self.connection, 
            max_identifier_length=128, 
            pool_size=10, 
            max_overflow=0,
            # connect_args={'application_name': 'Senac'}
        )

    def get_engine(self):
        """
            Obtém a sessão

            RETURN:
                Retorna a engine
        """

        return self.engine

    def create_session(self):
        """
            Cria uma sessão.

            RETURN:
                Retorna a sessão criada.
        """

        try:
            self.Session = sessionmaker(self.engine)
            self.session = self.Session()
            return self.session
        except Exception as error:
            print(f"ERROR: funcion {Connection.create_session.__name__} -> error -> {str(error)}")
            return str(error)

    def close_engine(self):
        """
            Fecha a engine

            RETURN:
                Fecha a engine
        """

        try:
            self.engine.dispose()
        except Exception as error:
            print(f"ERROR: funcion {Connection.close_engine.__name__} -> error -> {str(error)}")
    
    def create_database(self, db_name: str):
        """
            Função responsavel por criar a database/collection caso não exista
            
            PARAMETROS:
                nome do database/collection
        """
        with conn.engine.connect() as connection:
            try:
                # Read a single record
                sql = f"CREATE DATABASE IF NOT EXISTS {db_name};"
                connection.execute(sql)
                print(f"Database {db_name} criada com sucesso")
            except Exception as error:
                print("Erro ao criar base de dados", error)
    
    def drop_tables(self, tables: list):
        """
            Função responsavel por dropar as tabelas caso exista

            PARAMETROS:
                Lista de nomes das tabelas
        """
        with conn.engine.connect() as connection:
            try:
                for table_name in tables:
                    # Read a single record
                    sql = f"DROP TABLE  IF EXISTS {table_name};"
                    connection.execute(sql)
                    print(f"TABLE {table_name} dropada com sucesso")
            except Exception as error:
                print("Erro ao criar base de dados", error)        
```

Mapeando as entidades para Object Relational Mapper (ORM) 

```python

class Diciplina(Base):
    
    __tablename__ = 'tb_diciplina'
    __table_args__ = {'schema': 'senac'}

    id = Column(Integer, primary_key=True, autoincrement=True) 
    nome = Column(String(255))
    ch = Column(Integer)
    cp = Column(Integer)
    cr = Column(Integer)

    def workload_data_from_dataframe(self, df_diciplina: pd.DataFrame):
        """
            Função responsavel por realizar a carga na tabela de diciplinas
            
            PARÂMETROS:
                df_diciplina: dataframe de diciplinas;
        """
        try:
            df_diciplina.to_sql(
                name='tb_diciplina',
                con=conn.get_engine(),
                schema ='senac',
                if_exists='append',
                index=False,
                chunksize=10000,
                # dtype=validação de tipagem de dado, 
                method='multi'
            )
        except Exception as error:
            print("Error ao inserir carga de diciplinas", error)    

class Docente(Base):
    
    __tablename__ = 'tb_docente'
    __table_args__ = {'schema': 'senac'}

    id = Column(Integer, primary_key=True, autoincrement=True) 
    nome = Column(String(255))
    email = Column(String(50), nullable=False)
    celular = Column(String(20), nullable=False)

    def workload_data_from_dataframe(self, df_docente: pd.DataFrame):
        """
            Função responsavel por realizar a carga na tabela de docentes
            PARÂMETROS:
                df_docente: dataframe de docentes;
        """
        try:
            df_docente.to_sql(
                name='tb_docente', # nome da tabela
                con=conn.get_engine(), # engine da minha conexão
                schema ='senac', # nome do schema ou database
                if_exists='append', # se a tabela existir, ele apenas insere os registros
                index=False, # não incluir o index do dataframe
                chunksize=10000, # quantida de linhas que cada batch irá inserir por vez
                # dtype=validação de tipagem de dado, 
                method='multi' # passa diversos valores em uma única chamada
            )
        except Exception as error:
            print("Error ao inserir carga de docentes", error)
```
